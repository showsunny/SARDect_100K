```python
import json
import os
import numpy as np
import cv2
from tqdm import tqdm
import shutil
import matplotlib.pyplot as plt
import random
from collections import defaultdict

from SARDet_100K.bbox.utils import str2annot, draw_bboxes, load_image
from SARDet_100K.bbox.utils import clip_bbox, coco2voc, voc2yolo
```

```python
# 定义目录路径
base_dir = '/content/SARDet_100K/JPEGImages'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

# 确保目标目录存在
if not os.path.exists(base_dir):
    os.makedirs(base_dir)

# 定义一个函数，用于移动文件
def move_files(source_dir, target_dir):
    for filename in os.listdir(source_dir):
        source_file = os.path.join(source_dir, filename)
        target_file = os.path.join(target_dir, filename)
        if os.path.isfile(source_file):
            shutil.move(source_file, target_file)

# 移动文件并删除原始文件夹
move_files(train_dir, base_dir)
shutil.rmtree(train_dir)
move_files(val_dir, base_dir)
shutil.rmtree(val_dir)
move_files(test_dir, base_dir)
shutil.rmtree(test_dir)
print("All images have been moved and the folders have been deleted.")
```

```python
# 定义文件路径
file_path = '/content/SARDet_100K/Annotations/test.json'
# 打开并加载JSON文件
with open(file_path, 'r') as file:
    data = json.load(file)


# 查看文件内容的顶层结构
print("Keys in the JSON file:", data.keys())
# 查看图像信息
print("Sample image info:", data['images'][128])
# 查看标注信息
print("Sample annotation info:", data['annotations'][128])
# 查看类别信息
print("Sample category info:", data['categories'])
```

```python
# 加载数据
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# 统计类别实例数量
def count_instances(data):
    # 创建一个默认字典，存储每个类别的计数
    category_count = defaultdict(int)

    # 遍历所有注释，统计每个类别的实例数量
    for annotation in data['annotations']:
        category_id = annotation['category_id']
        category_count[category_id] += 1

    return category_count

# 打印类别实例数量
def print_instance_counts(category_count, category_map, dataset_name):
    print(f"\n{dataset_name} Category Instance Counts:")
    for category_id, count in category_count.items():
        category_name = category_map.get(category_id, "Unknown")
        print(f"{category_name}: {count}")

# 绘制类别实例数量分布图
def plot_category_distribution(train_data, val_data, categories):
    # 获取训练集和验证集的类别计数
    train_count = count_instances(train_data)
    val_count = count_instances(val_data)

    # 获取类别ID和名称的映射
    category_map = {category['id']: category['name'] for category in categories}

    # 打印每个数据集的类别实例数量
    print_instance_counts(train_count, category_map, "Training Set")
    print_instance_counts(val_count, category_map, "Validation Set")

    # 准备绘图数据
    category_ids = list(category_map.keys())
    train_counts = [train_count.get(cat_id, 0) for cat_id in category_ids]
    val_counts = [val_count.get(cat_id, 0) for cat_id in category_ids]
    category_names = [category_map[cat_id] for cat_id in category_ids]

    # 绘制图形
    x = range(len(category_ids))
    plt.figure(figsize=(12, 6))
    plt.bar(x, train_counts, width=0.4, label='Train', align='center')
    plt.bar(x, val_counts, width=0.4, label='Validation', align='edge')

    # 设置图形标签
    plt.xlabel('Categories')
    plt.ylabel('Instance Count')
    plt.title('Category Distribution in Train and Validation Sets')
    plt.xticks(x, category_names, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()

    # 显示图形
    plt.show()

# 文件路径
train_file = '/content/SARDet_100K/Annotations/train.json'
val_file = '/content/SARDet_100K/Annotations/val.json'

# 加载训练集和验证集数据
train_data = load_data(train_file)
val_data = load_data(val_file)

# 获取类别信息
categories = train_data['categories']

# 绘制类别分布图
plot_category_distribution(train_data, val_data, categories)
```

```python
# 文件路径
train_json_file = '/content/SARDet_100K/Annotations/train.json'
val_json_file = '/content/SARDet_100K/Annotations/val.json'
test_json_file = '/content/SARDet_100K/Annotations/test.json'
images_dir = '/content/SARDet_100K/JPEGImages/'
output_dir = '/content/SARDet_100K/YOLOv8_data/'

# 创建输出目录
os.makedirs(os.path.join(output_dir, 'images/train'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'images/val'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'images/test'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'labels/train'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'labels/val'), exist_ok=True)
os.makedirs(os.path.join(output_dir, 'labels/test'), exist_ok=True)

# 加载JSON数据
def load_data(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

train_data = load_data(train_json_file)
val_data = load_data(val_json_file)
test_data = load_data(test_json_file)

# 类别映射
category_map = {category['id']: category['name'] for category in train_data['categories']}

# 转换标注并保存到YOLO格式
def convert_annotations_to_yolo(data, split='train'):
    # 获取图片信息
    images = {image['id']: image for image in data['images']}

    # 遍历所有图片
    for image_info in tqdm(images.values()):
        image_file = image_info['file_name']
        image_width = image_info['width']
        image_height = image_info['height']

        # 获取图片路径
        image_path = os.path.join(images_dir, image_file)

        # 获取当前图片的标注
        annotations_for_image = [annotation for annotation in data['annotations'] if annotation['image_id'] == image_info['id']]

        # 如果当前图片有标注，则创建标签文件
        if annotations_for_image:
            new_image_path = os.path.join(output_dir, f'images/{split}', image_file)
            shutil.move(image_path, new_image_path)  # 使用shutil.move移动文件

            # 创建标注文件
            label_file = image_file.replace(os.path.splitext(image_file)[-1], '.txt')
            label_file_path = os.path.join(output_dir, f'labels/{split}', label_file)

            with open(label_file_path, 'w') as label_file:
                for annotation in annotations_for_image:
                    category_id = annotation['category_id']
                    coco_bbox = np.array(annotation['bbox'])
                    bbox = coco2voc(coco_bbox, image_height, image_width)
                    bbox = clip_bbox(bbox, image_height, image_width)
                    # 转换bbox为YOLO格式
                    x_center, y_center, width_ratio, height_ratio = voc2yolo(bbox, image_height, image_width)
                    # 写入YOLO格式标注
                    label_file.write(f'{category_id} {x_center:.6f} {y_center:.6f} {width_ratio:.6f} {height_ratio:.6f}\n')

# 转换训练集、验证集和测试集的标注
convert_annotations_to_yolo(train_data, 'train')
convert_annotations_to_yolo(val_data, 'val')
convert_annotations_to_yolo(test_data, 'test')

print("\n\nData conversion to YOLO format is complete.")
```

```python
# 类别列表

image_dir = '/content/SARDet_100K/YOLOv8_data/images/train/'
label_dir = '/content/SARDet_100K/YOLOv8_data/labels/train/'

y = 3; x = 4
plt.figure(figsize=(3.2*x, 3.2*y))
for idx in range(x*y):
    random_file = random.choice(os.listdir(image_dir))
    img = load_image(os.path.join(image_dir, random_file))
    image_height, image_width = img.shape[:2]
    label_file =os.path.splitext(random_file)[0]+ '.txt'
    with open(os.path.join(label_dir, label_file)) as f:
        annot = str2annot(f.read())
    bboxes_yolo = annot[...,1:]
    labels      = annot[..., 0].astype(int).tolist()
    names       = [category_map[label] for label in labels]
    plt.subplot(y, x, idx+1)
    plt.imshow(draw_bboxes(img = img,
                           bboxes = bboxes_yolo,
                           classes = names,
                           class_ids = labels,
                           class_name = True,
                           colors = colors,
                           bbox_format = 'yolo',
                           line_thickness = 2))
    plt.axis('OFF')
plt.tight_layout()
plt.show()
# 图像和标签路径
# image_path = "/content/SARDet_100K/YOLOv8_data/images/train/0063792.jpg"
# label_path = "/content/SARDet_100K/YOLOv8_data/labels/train/0063792.txt"


```
